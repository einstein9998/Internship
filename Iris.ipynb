```python
def nearestNeighborsTest(dataSet, targets, trainingSplit=0.7, n_neighbors=15, metric='minkowski', seed=1435):
    size = len(dataSet)
    num_features = len(dataSet[0])
    
    #add a target column so that the type of iris is preserved when the data is randomized
    targetAndData = np.append(targets[..., None], dataSet, 1)
    
    #choose some of the indices for the training group, and the remaining for the test group
    np.random.seed(seed)
    a = np.arange(size)
    np.random.shuffle(a)
    split = int(trainingSplit * size)
    trainingDataIndices = a[:split]
    queryDataIndices = a[split:]
    
    #build the data sets from the master set
    trainingDataWithTarget = targetAndData[trainingDataIndices, :]
    queryDataWithTarget = targetAndData[queryDataIndices, :]
    
    #cut off the target column
    trainingData = trainingDataWithTarget[:, range(1, num_features + 1)]
    queryData = queryDataWithTarget[:, range(1, num_features + 1)]
    
    #create a nerest neighbors model trained on the training data
    m = NearestNeighbors(n_neighbors=n_neighbors, metric=metric)
    m.fit(trainingData)
    
    #pass in the queries
    distances, indices = m.kneighbors(queryData)

    #Query number, Query label, Closest neighbor's label, Closest distance
    ret = [[i, int(queryDataWithTarget[i][0]),  int(trainingDataWithTarget[indices[:, 0][i]][0]) ,distances[:, 0][i]] for i in range(len(indices[:, 0]))]

    success_rate = sum(ret[i][1] == ret[i][2] for i in range(len(ret)))/len(ret)
    
   #print("Success Rate: " + str(success_rate * 100) + "%")
    
    
    return ret, success_rate
    
```

```python
#Let's try varying how much of the data is used for training

x = []
y = []
for split in np.arange(0.1, 1.0, 0.05):
    _, success = nearestNeighborsTest(iris.data, iris.target, split)
    x.append(split)
    y.append(success)
    
plt.figure(2)
plt.xlabel("Proportion of Data Used to Train")
plt.ylabel("Classification Success Rate")
plt.title("Training Data Size vs Success Rate with Minkowski")
plt.scatter(x, y)
# plt.text(0.15, 0.925, "Spicy Data Point", fontsize=7)
# plt.text(0.18, 0.945, "Slightly Helpful Data Point", fontsize=7)
# plt.text(0.53, 0.988, "Evil Data Points", fontsize=7)
# plt.text(0.65, 1.005, "Savior Data Point", fontsize=7)
plt.show()
```

```python
#Let's explore the efficacy of the different distance metrics
x=[]
y=[]

metrics =  ['cityblock', 'cosine', 'euclidean', 'l1', 'l2', 'manhattan', 'braycurtis', 'canberra', 'chebyshev', 'correlation', 'dice', 'hamming', 'jaccard', 'kulsinski', 'matching', 'minkowski', 'rogerstanimoto', 'russellrao', 'sokalmichener', 'sokalsneath', 'sqeuclidean']
for metric in metrics:
    _, success = nearestNeighborsTest(iris.data, iris.target, metric=metric)
    x.append(metric)
    y.append(success)
    
plt.figure(3)
plt.xlabel("Distance Metric")
plt.xticks(rotation=90)
plt.ylabel("Classification Success Rate")
plt.title("Metric vs Success Rate with 70% Training Data")
plt.scatter(x, y)
plt.show()
    
```

```python
#We conclude that the metrics dice, jaccard, kulsinski, matching, 
#rogerstanimoto, sokalmichener, and sokalsneath are unsuitable for this type of data
```
