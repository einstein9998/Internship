```python
import numpy as np
```

```python
binRands = np.random.binomial(1000, 0.3, 1000)
geoRands = np.random.geometric(0.2, 1000)
expRands = np.random.exponential(1.0, 1000)
gammaRands = np.random.gamma(0.2, 1.0, 1000)
hypgeoRands = np.random.hypergeometric(1000, 15000, 1000, 1000)
lRands = np.random.laplace(0.0, 1.0, 1000)
logRands = np.random.logistic(0.0, 1.0, 1000)
lognormRands = np.random.lognormal(0.0, 1.0, 1000)
poissonRands = np.random.poisson(166.6, 1000)
cauchyRands = np.random.standard_cauchy(1000)
triangleRands = np.random.triangular(0, 50, 100, 1000)
uniformRands = np.random.uniform(0, 100, 1000)
powerRands = np.random.power(1.44, 1000)
fRands = np.random.f(50, 50, 1000)
rlRands = np.random.rayleigh(1.0, 1000)
```

```python
from bokeh.layouts import row
from bokeh.plotting import figure, show

np.random.seed(234312)

hist1, edges1 = np.histogram(binRands, bins=50)
hist2, edges2 = np.histogram(geoRands, bins=50)
hist3, edges3 = np.histogram(expRands, bins=50)
hist4, edges4 = np.histogram(gammaRands, bins=50)
hist5, edges5 = np.histogram(hypgeoRands, bins=50)
hist6, edges6 = np.histogram(lRands, bins=50)
hist7, edges7 = np.histogram(logRands, bins=50)
hist8, edges8 = np.histogram(lognormRands, bins=50)
hist9, edges9 = np.histogram(poissonRands, bins=50)
hist10, edges10 = np.histogram(cauchyRands, bins=50)
hist11, edges11 = np.histogram(triangleRands, bins=50)
hist12, edges12 = np.histogram(uniformRands, bins=50)
hist13, edges13 = np.histogram(powerRands, bins=50)
hist14, edges14 = np.histogram(fRands, bins=50)
hist15, edges15 = np.histogram(rlRands, bins=50)

p1 = figure()
p1.quad(top=hist1, bottom=0, left=edges1[:-1], right=edges1[1:])

p2 = figure()
p2.quad(top=hist2, bottom=0, left=edges2[:-1], right=edges2[1:])

p3 = figure()
p3.quad(top=hist3, bottom=0, left=edges3[:-1], right=edges3[1:])

p4 = figure()
p4.quad(top=hist4, bottom=0, left=edges4[:-1], right=edges4[1:])

p5 = figure()
p5.quad(top=hist5, bottom=0, left=edges5[:-1], right=edges5[1:])

p6 = figure()
p6.quad(top=hist6, bottom=0, left=edges6[:-1], right=edges6[1:])

p7 = figure()
p7.quad(top=hist7, bottom=0, left=edges7[:-1], right=edges7[1:])

p8 = figure()
p8.quad(top=hist8, bottom=0, left=edges8[:-1], right=edges8[1:])

p9 = figure()
p9.quad(top=hist9, bottom=0, left=edges9[:-1], right=edges9[1:])

p10 = figure()
p10.quad(top=hist10, bottom=0, left=edges10[:-1], right=edges10[1:])

p11 = figure()
p11.quad(top=hist11, bottom=0, left=edges11[:-1], right=edges11[1:])

p12 = figure()
p12.quad(top=hist12, bottom=0, left=edges12[:-1], right=edges12[1:])

p13 = figure()
p13.quad(top=hist13, bottom=0, left=edges13[:-1], right=edges13[1:])

p14 = figure()
p14.quad(top=hist14, bottom=0, left=edges14[:-1], right=edges14[1:])

p15 = figure()
p15.quad(top=hist15, bottom=0, left=edges15[:-1], right=edges15[1:])


p = row(p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, p11, p12, p13, p14, p15)

show(p)
```
