```python
def knn(rest, sample_from, success, n_neighbors, algorithm, metric, seed):
    size = int(sample_from.shape[0] * 0.2 + .5)
    #print(size)
    query = sample_from.sample(n=size, random_state=seed)
    sub = list(set(sample_from.index.values) - set(query.index.values))
    act_without_query = sample_from.loc[sub]
    database = pd.concat([act_without_query, egfr_inactive_features, rest])
    
    model = neighbors.NearestNeighbors(n_neighbors=n_neighbors, algorithm=algorithm, metric=metric)
    start = time.time()
    model.fit(database)
    end = time.time()
    time_to_build = end - start
    start = time.time()
    _, indices = model.kneighbors(query)
    #print(indices[:10])
    #print(distances[:10])
    end = time.time()
    time_to_query = end - start
    sum = 0
    for i in indices:
        count = 0
        for j in i:
            if database.index.values[j] in success.index.values:
                count += 1
        sum += count/float(n_neighbors)
    average = sum/query.shape[0]
    d = OrderedDict({'algorithm':[algorithm], 'n_neighbors':[n_neighbors], 'metric':[metric], 
                              'build time (s)':[time_to_build], 'query time (s)':[time_to_query], 'quality':[average]})
    
    return pd.DataFrame(data=d), indices, distances
```

```python
stuff = []
for algorithm in ['brute', 'kd_tree', 'ball_tree']:
    for n in [50, 100]:
        stuff.append(knn(selection, egfr_active_features, egfr_active, n, algorithm, 'euclidean', 0))

pd.concat(i[0] for i in stuff)
```

```python
stuff2 = []
for n in range(5, 100, 5):
    stuff2.append(knn(selection, egfr_active_features, egfr_active, n, 'brute', 'euclidean', 0))

a = pd.concat(i[0] for i in stuff2)
```
